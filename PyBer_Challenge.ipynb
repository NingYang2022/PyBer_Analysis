### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
# Display city_data_df
city_data_df
```

```python
# Display ride_data_df
ride_data_df
```

### Merge the DataFrames

```python
# Combine the data into a single dataset

# When we merge two DataFrames, 
# we merge on a column with the same data, and the same column name
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = total_fares_by_type / total_rides_by_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fares_by_type / total_drivers_by_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 

# use column-oriented way, convert a dictionary of lists to a DF
pyber_summary_df = pd.DataFrame({
    "Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fares_by_type,
    "Average Fare per Ride": average_fare_per_ride,
    "Average Fare per Driver": average_fare_per_driver
    })

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

# Apply formatting to the following 3 columns:
# 1. "Total Fares"
# 2. "Average Fare per Ride"
# 3. "Average Fare per Driver"

# Add "$" signs and convert the amount to 2 decimal places
# Apply a thousands seprator to the data
# Apply the format to the 3 columns 
#   by chaining the map() and format() functions

pyber_summary_df["Total Fares"]= \
    pyber_summary_df["Total Fares"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Ride"]=\
    pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Driver"]=\
    pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares.
#  for each date where the indices are the city type and date.

# use groupby() on "type" and "date" columns, 
# apply sum() method on "fare" column.
# use [["fare"]] to create a DataFrame instead of a Series
sum_fares_per_type_per_date_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]

sum_fares_per_type_per_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. 
# This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fares_per_type_per_date_df = sum_fares_per_type_per_date_df.reset_index()

sum_fares_per_type_per_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, 
#  the columns ='type', and values='fare' 
# to make each cell have the total fares for each type of city by the date. 

sum_fares_per_type_per_date_pivot_df = \
    sum_fares_per_type_per_date_df.pivot(index="date", columns="type", values="fare")

sum_fares_per_type_per_date_pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame 
# using loc on the given dates, '2019-01-01':'2019-04-29'.
# the end date is not included, whitch means date range
# is 2019-01-01 through 2019-04-28

sum_fares_per_type_per_date_pivot_df = \
        sum_fares_per_type_per_date_pivot_df.loc['2019-01-01':'2019-04-29']

sum_fares_per_type_per_date_pivot_df
```

```python
# 6. Set the "date" index to datetime datatype,
# This is necessary to use the resample() method in Step 8.
sum_fares_per_type_per_date_pivot_df.index = \
    pd.to_datetime(sum_fares_per_type_per_date_pivot_df.index)

sum_fares_per_type_per_date_pivot_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_fares_per_type_per_date_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' 
# and get the sum of the fares for each week.
sum_fares_per_type_per_week_df = \
    sum_fares_per_type_per_date_pivot_df.resample('W').sum()

sum_fares_per_type_per_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Use the fig style fivethirtyeight.

plt.style.use('fivethirtyeight')


# Plot the resampled DataFrame using the df.plot() function.
# Set axis lable font size to 11
fig = sum_fares_per_type_per_week_df.plot(
                figsize=(15, 5),
                title="Total Fare by City Type",
                fontsize="11",
                linewidth = 1.3, 
                )


# set ylabel's title and font size.
fig.set_ylabel('Fare($USD)', fontsize=12)

# Remove xlable's title
fig.xaxis.label.set_visible(False)

# Set title font size to 12
fig.axes.title.set_size(12)

# Set legent font size of items to 10, font size of title to 10, and location to "center",  
fig.legend( fontsize=10,
            loc="center",
            title="Type",
            title_fontsize=10)

# Save the figure before display 
plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()





```
